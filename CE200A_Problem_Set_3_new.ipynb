### Import relevant packages

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
```

### Define the domain for the flow

```python
# # Create the x- and y-grid over the domain
# # Note the arguments: np.arange(min value, max value, interval)
# xx = np.arange(-4,4,0.2)
# yy = np.arange(-4,4,0.2)

# # Create the grid using the meshgrid function
# x, y = np.meshgrid(xx, yy)

# print(x)
# print(y)
```

```python
xlim, ylim = (-5, 10), (-5, 5)  # Limits for the plot

# Create a grid of x, y points
xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
x, y = np.meshgrid(xx, yy)
```

### Example: Define the parameters for the Couette flow

```python
U = 1
m = 5
u = U + (m / (2 * np.pi)) * x / (x**2 + y**2)
v = (m / (2 * np.pi)) * y / (x**2 + y**2)

```

### Plot a vertical profile of the flow

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(u[:,2],y[:,2],'.-') #extract the values at a specific x location
ax.set_xlabel('u')
ax.set_ylabel('y')
ax.set_title('Velocity field - vertical profile')
```

### **Note**: Be sure to add axis labels and titles to all your plots, as above

### Create the quiver plot

```python
# Create a quiver plot
fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(x,y,u,v)
ax.set_aspect('equal')
```

```python
# Make the quiver plot more colorful
fig, ax = plt.subplots(figsize=(7,7))
speed = np.sqrt(u**2+v**2)
ax.quiver(x,y,u,v, speed, alpha=0.8)
ax.set_aspect('equal')
```

### Make a contour plot

```python
fig, ax = plt.subplots(figsize=(7,7))
CS = ax.pcolor(x, y, speed)
ax.set_aspect('equal')
fig.colorbar(CS)
```

### Plot the streamlines

```python
# Create streamline plot
fig, ax = plt.subplots(figsize=(7,7))
ax.streamplot(x, y, u, v, color=speed, cmap = "winter")
```
